#### 1. Напишите код, который будет переводить полярные координаты в декартовы.

```python
import numpy as np

def pol_in_dec_coord(r:float, corner:int):
    """ переводит полярные координаты в декартовые """
    x = r*np.cos(radians(corner))
    y = r*np.sin(radians(corner))
    
    print(f'x = {x}, y = {y}')

pol_in_dec_coord(9, 180)
```

##### 1 Система из 2-х уравнений
exp(x) + x∙(1 – y) = 1

```python
# Строим графики уравнений:

x = np.linspace(-6, 6, 200)
y = x**2 - 1

plt.plot(x, y, label=f'$y = x^2 – 1$')

y = (((np.exp(x)) - 1) / x) + 1
plt.plot(x, y, label=f'$exp(x) + x∙(1 – y) = 1$')

plt.legend()
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

#### 2. Решите систему уравнений и неравенств:

y = x2 – 1

exp(x) + x∙(1 – y)  - 1 > 0


```python
x = np.linspace(-3, 5, 200)

y = np.exp(x) - 1 + x*(2-x**2)
plt.plot(x, y)
plt.grid(True)
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from scipy.optimize import fsolve

def equations(x):
    return np.exp(x) - 1 + x*(2-x**2)

print(f'x_1:{fsolve(equations, -1.5)[0]}\t y_1: {(fsolve(equations, -1.5)[0])**2 - 1}\n',\
      f'x_2:{fsolve(equations, 0)[0]}\t\t y_2: {(fsolve(equations, 0)[0])**2 - 1}\n',\
      f'x_3:{fsolve(equations, 2.5)[0]}\t y_3: {(fsolve(equations, 2.5)[0])**2 - 1}\n',\
      f'x_4:{fsolve(equations, 4)[0]}\t y_4: {(fsolve(equations, 4)[0])**2 - 1}'
     )
```
