#### Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. Например (10, 10, 10):

```python
%matplotlib inline
import matplotlib as plt
import math
def vector_len(x1, x2, x3):
    v_l = math.sqrt(x1**2 + x2**2 + x3**2) 
    print(f'Длина вектора = ', v_l)
vector_len(10, 10, 10)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
import numpy as np
x = np.linspace(-5, 5, 21)
y = 3*x + 1
y2 = -1/3*x - 1/3
plt.plot(x, y)
plt.plot(x, y2)
plt.xlabel("x")
plt.ylabel("y")
```

```python
x = np.linspace(-5,5,201)
y1 = np.sqrt((4 - x**2))
y2 = -(np.sqrt(4 - x**2))
plt.axis('scaled')
plt.ylim(-5, 5)
plt.xlim(-5, 5)

plt.grid(True)
plt.plot(x, y1)
plt.plot(x, y2)
```

#### 3.2 График гиперболы:

```python
import matplotlib.pyplot as plt
import decimal
import numpy as np

xmin = -20
xmax = 20
dx = 0.1

xlist = np.around(np.arange(xmin, xmax, dx), decimals=4)

ylist = 1 / xlist
plt.grid(True)
plt.plot(xlist, ylist)
plt.show()
```

##### 1. Нарисуйте трехмерный график двух параллельных плоскостей:

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-50, 50, 1)
Y = np.arange(-50, 50, 1)
X, Y = np.meshgrid(X, Y)
Z2 = 1 - 2*X - 3*Y
Z1 = 48 - 2*X - 3*Y
ax.plot_surface(X, Y, Z1)
ax.plot_surface(X, Y, Z2)
show()
```

#### 2. Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.axes(projection='3d')

def f1(x, y): 
    return np.sqrt((y**2/90 + x**2/100)/2)

def f2(x, y): 
    return -np.sqrt((y**2/90 + x**2/100)/2)

x = np.linspace(-100, 100, 10) 
y = np.linspace(-100, 100, 10) 
   
X, Y = np.meshgrid(x, y) 
Z1 = f1(X, Y)
Z2 = f2(X, Y)
 
ax = plt.axes(projection ='3d') 
# ax.plot_wireframe(X, Y, Z1, color ='blue')
# ax.plot_wireframe(X, Y, Z2, color ='red')
ax.plot_surface(X, Y, Z1, rstride=1, cstride=1,
                cmap='viridis')
ax.plot_surface(X, Y, Z2, rstride=1, cstride=1,
                cmap='brg')

plt.show()
```
